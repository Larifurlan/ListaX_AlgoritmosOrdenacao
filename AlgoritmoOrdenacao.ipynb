## Bubble Sort

```python
import random
import time
import statistics

random.seed(42)

vetor_1000 = [random.randint(1, 1000000) for i in range(1000)]
vetor_10000 = [random.randint(1, 1000000) for i in range(10000)]

def bubble_sort(lista):
    vetor = lista[:]
    trocas = 0
    n = len(vetor)
    for i in range(n):
        for j in range(n - i - 1):
            if vetor[j] > vetor[j + 1]:
                vetor[j], vetor[j + 1] = vetor[j + 1], vetor[j]
                trocas = trocas + 1
    return vetor, trocas

for vetor in [vetor_1000, vetor_10000]:
    tempos = []
    for execucao in range(3):
        inicio = time.perf_counter()
        resultado, ops = bubble_sort(vetor)
        fim = time.perf_counter()
        tempos.append(round(fim - inicio, 6))
    media = round(statistics.mean(tempos), 6)
    desvio = round(statistics.stdev(tempos), 6)
    print(f"n={len(vetor)} | exec1={tempos[0]} | exec2={tempos[1]} | exec3={tempos[2]} | media={media} | desvio={desvio} | trocas={ops}")

print("n=100000 | N/C")
```

## Merge Sort

```python
import random
import time
import statistics

random.seed(42)

vetor_1000 = [random.randint(1, 1000000) for i in range(1000)]
vetor_10000 = [random.randint(1, 1000000) for i in range(10000)]
vetor_100000 = [random.randint(1, 1000000) for i in range(100000)]

def merge_sort(lista):
    movimentos = [0]

    def dividir(vetor):
        if len(vetor) <= 1:
            return vetor
        meio = len(vetor) // 2
        esquerda = dividir(vetor[:meio])
        direita = dividir(vetor[meio:])
        return juntar(esquerda, direita)

    def juntar(esquerda, direita):
        resultado = []
        i = 0
        j = 0
        while i < len(esquerda) and j < len(direita):
            if esquerda[i] <= direita[j]:
                resultado.append(esquerda[i])
                i = i + 1
            else:
                resultado.append(direita[j])
                j = j + 1
            movimentos[0] = movimentos[0] + 1
        while i < len(esquerda):
            resultado.append(esquerda[i])
            i = i + 1
            movimentos[0] = movimentos[0] + 1
        while j < len(direita):
            resultado.append(direita[j])
            j = j + 1
            movimentos[0] = movimentos[0] + 1
        return resultado

    vetor = lista[:]
    vetor = dividir(vetor)
    return vetor, movimentos[0]

for vetor in [vetor_1000, vetor_10000, vetor_100000]:
    tempos = []
    for execucao in range(3):
        inicio = time.perf_counter()
        resultado, ops = merge_sort(vetor)
        fim = time.perf_counter()
        tempos.append(round(fim - inicio, 6))
    media = round(statistics.mean(tempos), 6)
    desvio = round(statistics.stdev(tempos), 6)
    print(f"n={len(vetor)} | exec1={tempos[0]} | exec2={tempos[1]} | exec3={tempos[2]} | media={media} | desvio={desvio} | movimentos={ops}")
```

## Quick Sort

```python
import random
import time
import statistics
import sys
sys.setrecursionlimit(300000)

random.seed(42)

vetor_1000 = [random.randint(1, 1000000) for i in range(1000)]
vetor_10000 = [random.randint(1, 1000000) for i in range(10000)]
vetor_100000 = [random.randint(1, 1000000) for i in range(100000)]

def quick_sort(lista):
    trocas = [0]

    def ordenar(vetor, inicio, fim):
        if inicio < fim:
            pivo = vetor[fim]
            i = inicio - 1
            for j in range(inicio, fim):
                if vetor[j] <= pivo:
                    i = i + 1
                    vetor[i], vetor[j] = vetor[j], vetor[i]
                    trocas[0] = trocas[0] + 1
            vetor[i + 1], vetor[fim] = vetor[fim], vetor[i + 1]
            trocas[0] = trocas[0] + 1
            p = i + 1
            ordenar(vetor, inicio, p - 1)
            ordenar(vetor, p + 1, fim)

    vetor = lista[:]
    ordenar(vetor, 0, len(vetor) - 1)
    return vetor, trocas[0]

for vetor in [vetor_1000, vetor_10000, vetor_100000]:
    tempos = []
    for execucao in range(3):
        inicio = time.perf_counter()
        resultado, ops = quick_sort(vetor)
        fim = time.perf_counter()
        tempos.append(round(fim - inicio, 6))
    media = round(statistics.mean(tempos), 6)
    desvio = round(statistics.stdev(tempos), 6)
    print(f"n={len(vetor)} | exec1={tempos[0]} | exec2={tempos[1]} | exec3={tempos[2]} | media={media} | desvio={desvio} | trocas={ops}")
```
