## Bubble Sort

```python
import random
import time
import statistics

random.seed(42)

vetor_1000 = [random.randint(1, 1000000) for i in range(1000)]
vetor_10000 = [random.randint(1, 1000000) for i in range(10000)]

def bubble_sort(lista):
    vetor = lista[:]
    trocas = 0
    n = len(vetor)
    for i in range(n):
        for j in range(n - i - 1):
            if vetor[j] > vetor[j + 1]:
                vetor[j], vetor[j + 1] = vetor[j + 1], vetor[j]
                trocas = trocas + 1
    return vetor, trocas

for vetor in [vetor_1000, vetor_10000]:
    tempos = []
    for execucao in range(3):
        inicio = time.perf_counter()
        resultado, ops = bubble_sort(vetor)
        fim = time.perf_counter()
        tempos.append(round(fim - inicio, 6))
    media = round(statistics.mean(tempos), 6)
    desvio = round(statistics.stdev(tempos), 6)
    print(f"n={len(vetor)} | exec1={tempos[0]} | exec2={tempos[1]} | exec3={tempos[2]} | media={media} | desvio={desvio} | trocas={ops}")

print("n=100000 | N/C")
```
